```python
import requests
from bs4 import BeautifulSoup

# 요청(Request) 객체를 생성합니다.
request = requests.get('https://finance.naver.com/')

# 웹 사이트의 HTML 소스코드를 추출합니다.
html = request.text.strip()

# HTML 소스코드를 파이썬 BeautifulSoup 객체로 변환합니다.
soup = BeautifulSoup(html, 'html.parser')

# 특정한 클래스 이름으로 접근합니다.
data = soup.find('div',class_="section_strategy")

# <td> 태그 내부에서 <span> 태그를 포함하는 경우를 추출합니다.
data = data.select('span > a')

# 모든 링크에 하나씩 접근합니다.
for link in data:
    # 텍스트(글 제목) 부분만 출력합니다.
    print(link.text)



'''
title = soup.findAll("span", class_="pcol2 _setTop _setTopListUrl")

for ch in title:
    print(ch)
    print(ch.get_text())
'''
```

```python
import requests
from bs4 import BeautifulSoup

# 요청(Request) 객체를 생성합니다.
request = requests.get('https://finance.naver.com/news/news_read.naver?mode=mainnews&office_id=008&article_id=0004911653')

# 웹 사이트의 HTML 소스코드를 추출합니다.
html = request.text.strip()

# HTML 소스코드를 파이썬 BeautifulSoup 객체로 변환합니다.
soup = BeautifulSoup(html, 'html.parser')

# 특정한 클래스 이름으로 접근합니다.
data = soup.find('div',class_="u_cbox_text_wrap")

# 특정한 요소를 접근합니다.
data = soup.select('span')

# 모든 링크에 하나씩 접근합니다.
for link in data:
    # 텍스트(글 제목) 부분만 출력합니다.
    print(link.text)
```
