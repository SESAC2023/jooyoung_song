```python
#히스토그램 plot
import matplotlib.pyplot as plt
from matplotlib import font_manager

# 한국어 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # malgun.ttf는 윈도우 환경에서 기본으로 제공되는 한국어 폰트 경로입니다.
font_prop = font_manager.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font_prop.get_name()
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.family'] = 'Malgun Gothic' 

# 열 이름들 (선호도부터 기타까지)
column_names = df.columns[2:]

# 열 개수에 따라 그리드 설정
num_cols = len(column_names)
num_rows = 1

# 그래프 크기 조절
fig, axes = plt.subplots(num_cols, num_rows, figsize=(8, 12))

# 각 열별로 히스토그램 그리기
for i, col in enumerate(column_names):
    ax = axes[i]
    ax.hist(df[col].fillna(0), bins=20, alpha=0.7, color='blue', edgecolor='black')
    ax.set_title(col)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

### * 선호도 0, 1인 행에 대한 바이어스

```python
# 선호도 열에서 값이 0이나 1인 행만 남기기
df_filtered = df[(df['선호도'] == 0) | (df['선호도'] == 1)]

df_filtered.info()
```

```python
# 각 칼럼별 데이터의 개수 count

for col in df_filtered.columns[2:]:

    counts = df_filtered[col].value_counts()
    print(counts)
    print()
```

```python
# 각 칼럼별 데이터의 개수 count
for col in df_filtered.columns[2:]:
    counts = df_filtered[col].value_counts()
    
    # 제거할 값 리스트
    values_to_remove = [2.0, 3.0]
    
    # 제거할 값들이 있는 경우 해당 값들을 제외한 데이터만 필터링
    if any(value in counts.index for value in values_to_remove):
        df_filtered = df_filtered[df_filtered[col].isin(values_to_remove) == False]

# 필터링된 데이터프레임
# 각 칼럼별 데이터의 개수 count

for col in df_filtered.columns[2:]:

    counts = df_filtered[col].value_counts()
    print(counts)
    print()
```

```python
# 데이터 칼럼 이름과 데이터 개수
column_names = df_filtered.columns[4:]

# 플롯 크기 조절
plt.figure(figsize=(10, 8))

# 원형 히스토그램 그리기
plt.pie(df_filtered[column_names].sum(), labels=column_names, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # 원형을 유지하기 위해 설정
plt.title('Bias Distribution')

# 그래프 표시
plt.show()
```
